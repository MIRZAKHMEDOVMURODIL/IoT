# eqdataの読み込み・加工

```python
import glob
import timeit

start = timeit.default_timer()

path=r'C:\Users\00220401626\Desktop\General\05_グライディングセンタ\05_グライディングセンタ\eq_data\eq_data\20221108_091036\1\*.csv'
files=glob.glob(path)
df_list=[pd.read_csv(file, skiprows=2,encoding="cp932")for file in files]

for p in files:
    t = os.path.getmtime(p)
    D = datetime.datetime.fromtimestamp(t).strftime('%Y/%m/%d %H:%M:%S.%f')[:-3]
    
dx= pd.concat(df_list, axis=0,ignore_index=True)
dx = dx.rename(columns={'Time': 'Time(msec)', 'POSF': 'POSF(mm)','POSF.1': 'POSF.1(mm)','TCMD':'TCMD(%)','TCMD.1':'TCMD.1(%)','TCMD.2':'TCMD.2(%)','TCMD.3':'TCMD.3(%)','TCMD.4':'TCMD.4(%)','SPEED':'SPEED(1/min)'})
dx=dx.dropna()
dx["latest_time"]=D
dx["latest_time"]=pd.to_datetime(dx["latest_time"])
dx['Time(msec)'] = dx['Time(msec)'].astype(float).astype(int)
dx['Time(msec)']=dx['Time(msec)']/1000
dx['lt_unix']=dx.latest_time.apply(lambda t: int(t.timestamp()))
dx['lt_unix2']=dx['lt_unix']-299.995
dx['lt_unix3']=(dx['lt_unix2']+dx['Time(msec)']).astype(int)
dx['lt']=pd.to_datetime(dx['lt_unix3'], unit='s')
dx = dx[['lt','Time(msec)','POSF(mm)','POSF.1(mm)','TCMD(%)','TCMD.1(%)','TCMD.2(%)','TCMD.3(%)','TCMD.4(%)','SPEED(1/min)']]

stop = timeit.default_timer()
print('Time: ', stop - start)
```

```python

df=dx
#カラムの要素の変更
# df['TCMD(%)'] = df['TCMD(%)'].astype(float)
# df['TCMD.1(%)'] = df['TCMD.1(%)'].astype(float)
# df['TCMD.2(%)'] = df['TCMD.2(%)'].astype(float)
# df['TCMD.3(%)'] = df['TCMD.3(%)'].astype(float)
# df['TCMD.4(%)'] = df['TCMD.4(%)'].astype(float)
# df['SPEED(1/min)'] = df['SPEED(1/min)'].astype(float)
   
for col in df.columns[1:]:
    df[col] = df[col].astype(float)
    

#カラム名を変更
df = df.rename(columns={'TCMD(%)': 'X軸_TCMD(%)', 'TCMD.1(%)': 'Z軸_TCMD(%)','TCMD.2(%)':'B軸_TCMD(%)','TCMD.3(%)':'C軸_TCMD(%)','TCMD.4(%)':'スピンドル_TCMD(%)'})

#カラムの削除
df = df.drop(['Time(msec)','POSF(mm)','POSF.1(mm)'],axis=1)
```

```python
# 一定間隔で取り出し
step = 60
df_selected = df.loc[::step]
df_selected

# speedで絞る
time_list_s = []
time_list_g = []
# df_selected = df_selected.loc[df_selected['SPEED(1/min)'] > 2000]

for time,speed,speed2 in zip(df_selected['lt'], df_selected['SPEED(1/min)'],df_selected['SPEED(1/min)'].iloc[1:]):
    
    if (speed <= 2950) & (speed2 >= 3000.0):
        n = '加工開始',time
        # print(n)
        time_list_s.append(time)
        
    elif (speed >= 3000) & (speed2 <= 2950.0):
        n = '加工終了',time
        # print(n)
        time_list_g.append(time)



time_list_s2 = []
time_list_g2 = []
time_list = []


for s in time_list_s:
    for g in time_list_g:
        t = g - s
        
        if (t < datetime.timedelta(minutes=11)) & (t > datetime.timedelta(minutes=5)):
            if (s not in time_list_s2) & (g not in time_list_g2):
                time_list_s2.append(s)
                time_list_g2.append(g)
                time_list.append(t)

                
df_s = pd.DataFrame(time_list_s2, columns = ['加工開始時間'])
df_g = pd.DataFrame(time_list_g2, columns = ['加工終了時間'])
df_t = pd.DataFrame(time_list, columns = ['経過時間'])

df_all = pd.concat([df_s, df_g,df_t], axis=1)
# df_all = df_all.dropna()
df_all
```

# 3次元測定データの読み込み・加工

```python

dp=pd.DataFrame(columns=["フォルダー名","ファイル名","円錐角度_測定値","円錐角度_誤差値"])

for root, dirs, files in os.walk(r"C:\Users\00220401626\Desktop\General\05_グライディングセンタ\05_グライディングセンタ\三次元測定結果\1"):
    for f in files:
        full = os.path.join(root, f)
        
# #フォルダー読み込み
# #PDF読み取り
        doc = fitz.open(full)
#ファイル名を取得
        name = f
        Name = root
# １ページずつテキストを抽出して出力
        for page in range(len(doc)):
            text = doc[page].get_text()

#リスト化    
            Text = text.split('\n')

#PDFの中身を抽出
            for i in Text:
                if i =="円錐角度_円錐2":
                    Ti2=Text.index("円錐角度_円錐2")
                    data1=Text[Ti2+6]
                    data2=Text[Ti2+4]


#データフレーム化
                    df = pd.DataFrame([[Name,name,data1,data2]],columns=["フォルダー名","ファイル名","円錐角度_測定値","円錐角度_誤差値"])
                    dp=pd.concat([dp,df],axis=0,ignore_index=True)


dp['LotNo'] = dp['フォルダー名'].str[-11:]
dp=dp.loc[(dp["LotNo"] == '2218571-000')|(dp["LotNo"] == '2217682-000')|(dp["LotNo"] == '2217179-000')]

# '2217682-000'ver
dp1=dp.loc[dp["LotNo"] == '2217682-000']
dp1['memo']=''
dp1.loc[dp1['ファイル名'].astype(str).str.contains('No.'), 'memo'] = dp1['ファイル名'].astype(str).str.replace(r'(.*?)No.','',regex=True)
dp1['No'] = dp1['memo'].str[:-16]

# '2218571-000'ver
dp2=dp.loc[dp["LotNo"] == '2218571-000']
dp2['memo']=''
dp2.loc[dp2['ファイル名'].astype(str).str.contains('No.'), 'memo'] = dp2['ファイル名'].astype(str).str.replace(r'(.*?)No.','',regex=True)
dp2['No'] = dp2['memo'].str[:-16]

# '2217179-000'ver
dp3=dp.loc[dp["LotNo"] == '2217179-000']
dp3['memo']=''
dp3.loc[dp3['ファイル名'].astype(str).str.contains('No.'), 'memo'] = dp3['ファイル名'].astype(str).str.replace(r'(.*?)No.','',regex=True)
dp3['No'] = dp3['memo'].str[:-4]

dp=pd.concat([dp1,dp2,dp3],axis=0,ignore_index=True)
dp=dp[['フォルダー名','ファイル名','円錐角度_測定値','円錐角度_誤差値','LotNo','No']]
dp["ナンバリング"]=dp['LotNo']+"-"+dp['No']
print(dp.shape)
dp.head()
```

# PDFとtimedataの紐づけ

```python
import re
# データ整理
tp=pd.merge(dp ,tde ,how='left' ,on='ナンバリング')#dpとtdeが共通で持つ"ナンバリング"列を起点にして結合
# pd.set_option('display.max_rows', None)#作成したデータフレーム結果をすべて表示する#以降すべてに反映されるので注意！
# pd.set_option('display.max_columns', None)#作成したデータフレーム結果をすべて表示する#以降すべてに反映されるので注意！
tp=tp[['ファイル名','LotNo_x','No_x','円錐角度_測定値','円錐角度_誤差値','開始時間','終了時間']]
tp = tp.dropna(how='any', axis='index')
tp.reset_index(inplace=True, drop=True)


#誤差値を文字としてリスト化、カラム追加
gosa1=pd.DataFrame()
for i in tp['円錐角度_誤差値']:
    ddd = re.split('[°"]' , i)#　度と分秒に分けて抽出
    data1=ddd[0]
    data2=ddd[1]
#データフレーム化
    data3 = pd.DataFrame([[data1,data2]],columns=["度2","に2"])
    gosa1=pd.concat([gosa1,data3],axis=0,ignore_index=True)
tp=pd.concat([tp,gosa1],axis=1)


gosa2=pd.DataFrame()
for h in tp['に2']:
    ddd = re.split("[']" , h)#　分と秒を分ける
    data1=ddd[0]
    data2=ddd[1]
#データフレーム化
    data3 = pd.DataFrame([[data1,data2]],columns=["分2","秒2"])
    gosa2=pd.concat([gosa2,data3],axis=0,ignore_index=True)

tp=pd.concat([tp,gosa2],axis=1)

##int型に変換
# tp['度2'] = tp['度2'].astype(int)
tp['分2'] = tp['分2'].astype(int)
tp['秒2'] = tp['秒2'].astype(int)
# tp['誤差値(度)']=tp['分2']/60 + tp['秒2']/3600

for i in tp['度2']:
    if '-0' in i:
        tp.loc[tp['度2'] == i,'誤差値(度)'] = -(tp['分2']/60 + tp['秒2']/3600)
    else:
        tp.loc[tp['度2'] == i,'誤差値(度)'] = tp['分2']/60 + tp['秒2']/3600
print(tp.shape)
tp.head()
```

```python
#測定値を文字としてリスト化、カラム追加
real1=pd.DataFrame()
for i in tp['円錐角度_測定値']:
    ddd = re.split('[°"]' , i)
    data1=ddd[0]
    data2=ddd[1]
#データフレーム化
    df = pd.DataFrame([[data1,data2]],columns=["度1","に1"])#　度と分秒に分けて抽出
    real1=pd.concat([real1,df],axis=0,ignore_index=True)
tp=pd.concat([tp,real1],axis=1)


real2=pd.DataFrame()
for h in tp['に1']:
    ddd = re.split("[']" , h)
    data1=ddd[0]
    data2=ddd[1]
#データフレーム化
    df = pd.DataFrame([[data1,data2]],columns=["分1","秒1"])#　分と秒を分ける
    real2=pd.concat([real2,df],axis=0,ignore_index=True)
tp=pd.concat([tp,real2],axis=1)
    
##int型に変換
tp['度1'] = tp['度1'].astype(int)
tp['分1'] = tp['分1'].astype(int)
tp['秒1'] = tp['秒1'].astype(int)
tp['測定値(度)']=abs(tp['度1']) + tp['分1']/60 + tp['秒1']/3600
tp=tp[['ファイル名','LotNo_x','No_x','開始時間','終了時間','円錐角度_測定値','円錐角度_誤差値','誤差値(度)','測定値(度)']]
```

```python
for i in tp['誤差値(度)']:
    if i > 0.0333333333:
        tp.loc[tp['誤差値(度)'] == i, 'judge'] = 4
    if 0.022222222 < i <= 0.033333333: 
        tp.loc[tp['誤差値(度)'] == i, 'judge'] = 3
    if 0.011111111 < i <= 0.022222222: 
        tp.loc[tp['誤差値(度)'] == i, 'judge'] = 2
    if 0 < i <= 0.011111111: 
        tp.loc[tp['誤差値(度)'] == i, 'judge'] = 1
    if i == 0:
        tp.loc[tp['誤差値(度)'] == i, 'judge'] = 0
    if -0.011111111 <= i < 0:
        tp.loc[tp['誤差値(度)'] == i, 'judge'] = -1
    if -0.022222222 <= i < -0.011111111:
        tp.loc[tp['誤差値(度)'] == i, 'judge'] = -2
    if -0.033333333 <= i < -0.022222222:
        tp.loc[tp['誤差値(度)'] == i, 'judge'] = -3
    if i < -0.033333333:
        tp.loc[tp['誤差値(度)'] == i, 'judge'] = -4    

tp = tp.loc[ tp['LotNo_x'] != '2217682-000' ]
print(tp.shape)
tp.head()
```
