# eq_dataの読み込み
## d_lot：正味加工時間（lt）を付けた

```python
d0=pd.DataFrame()
for root, dirs, files in os.walk(r"C:\Users\00220401626\Desktop\General\05_グライディングセンタ\05_グライディングセンタ\eq_data\eq_data\20221108_091036\1"):
    for f in files:
        # 指定したフォルダ内のファイル情報をすべて取得
        full = os.path.join(root, f)        
        # ファイルの最終更新日時を取得
        p = full
        t = os.path.getmtime(p)
        D = datetime.datetime.fromtimestamp(t).strftime('%Y/%m/%d %H:%M:%S.%f')[:-3]#最終更新日時をタイムスタンプ型に変換
        
        # ファイル読み込み、整形
        dx=pd.read_csv(full, skiprows=2,encoding="cp932")
        dx = dx.rename(columns={'Time': 'Time(msec)','POSF': 'POSF(mm)','POSF.1': 'POSF.1(mm)','TCMD': 'X軸_TCMD(%)', 'TCMD.1': 'Z軸_TCMD(%)','TCMD.2':'B軸_TCMD(%)','TCMD.3':'C軸_TCMD(%)','TCMD.4':'スピンドル_TCMD(%)','SPEED':'SPEED(1/min)'})
        dx=dx.dropna()
        dx["latest_time"]=D
        dx["latest_time"]=pd.to_datetime(dx["latest_time"])#最終更新日時をdatetime型に変換
        dx['Time(msec)'] = dx['Time(msec)'].astype(float).astype(int)
        dx['Time(msec)']=dx['Time(msec)']/1000
        dx['lt_unix']=dx.latest_time.apply(lambda t: int(t.timestamp()))
        dx['lt_unix2']=dx['lt_unix']-299.995
        dx['lt_unix3']=(dx['lt_unix2']+dx['Time(msec)']).astype(int)
        dx['lt']=pd.to_datetime(dx['lt_unix3'], unit='s')

        # 縦結合
        d0=pd.concat([d0,dx], axis=0, ignore_index=True)
        # print(dx.dtypes.to_frame().rename(columns={0:'Column type'}))
d_lot = d0[['lt','Time(msec)','POSF(mm)','POSF.1(mm)','X軸_TCMD(%)','Z軸_TCMD(%)','B軸_TCMD(%)','C軸_TCMD(%)','スピンドル_TCMD(%)','SPEED(1/min)']]
print(d_lot.shape)
d_lot.head()
```

# [ 一定間隔で取り出し→グラフ化 ]
## df_selected: 一定間隔で取り出してdataframeに格納

# speedで絞る
## df_s : 加工開始時間
## df_g : 加工終了時間
## df_t : 経過時間
## df_all : 加工開始時間+加工終了時間+経過時間

```python
# 一定間隔で取り出し
step = 60
df_selected = df_lot.loc[::step]
df_selected

# speedで絞る
time_list_s = []
time_list_g = []
df_selected = df_selected.loc[df_selected['SPEED(1/min)'] > 2000]

for time,speed,speed2,speed3 in zip(df_selected['lt'], df_selected['SPEED(1/min)'],df_selected['SPEED(1/min)'].iloc[1:],df_selected['SPEED(1/min)'].iloc[2:]):
    
    if (speed <= 2500) & (speed2 >= 2700.0) & (speed3 >= 2700.0):
        n = '加工開始',time
        # print(n)
        time_list_s.append(time)
        
    elif (speed >= 2700) & (speed2 <= 2500.0) & (speed3 <= 2500.0):
        n = '加工終了',time
        # print(n)
        time_list_g.append(time)



time_list_s2 = []
time_list_g2 = []
time_list = []


for s in time_list_s:
    
    for g in time_list_g:
        t = g - s
        
        if t < datetime.timedelta(minutes=10) and t > datetime.timedelta(minutes=7):
            time_list_s2.append(s)
            time_list_g2.append(g)
            time_list.append(t)


df_s = pd.DataFrame(time_list_s2, columns = ['加工開始時間'])
df_g = pd.DataFrame(time_list_g2, columns = ['加工終了時間'])
df_t = pd.DataFrame(time_list, columns = ['経過時間'])

df_all = pd.concat([df_s, df_g,df_t], axis=1)
df_all = df_all.dropna()
df_all
```
