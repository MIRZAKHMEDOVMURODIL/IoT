# eq_data加工

```python
"""
eq_data : As-prerared eq_data
"""
eqPath = []

for root, dirs, files in os.walk(eqdata_Path):
    for f in files:
        # 指定したフォルダ内のファイル情報をすべて取得
        full = os.path.join(root, f)
        eqPath.append(full)
        # print(eqPath)
print('全ファイルパス取得完了')
print('現在、全eq_detaを縦結合しています。')

eq_data = pd.DataFrame()
for p in eqPath:    
    t = os.path.getmtime(p)
    D = datetime.datetime.fromtimestamp(t).strftime('%Y/%m/%d %H:%M:%S.%f')[:-3]#最終更新日時をタイムスタンプ型に変換
    
    # ファイル読み込み、整形
    d = pd.read_csv(p, skiprows=2,encoding="cp932")
    d = d.rename(columns={'Time': 'Time(msec)', 'POSF': 'POSF(mm)','POSF.1': 'POSF.1(mm)','TCMD':'TCMD(%)','TCMD.1':'TCMD.1(%)','TCMD.2':'TCMD.2(%)','TCMD.3':'TCMD.3(%)','TCMD.4':'TCMD.4(%)','SPEED':'SPEED(1/min)'})
    d = d.dropna()
    
    d["latest_time"] = D
    d["latest_time"] = pd.to_datetime(d["latest_time"])#最終更新日時をdatetime型に変換
    d['Time(msec)'] = (d['Time(msec)'].astype(float).astype(int))/1000 # d['Time(msec)'] = d['Time(msec)']/1000
    d['lt_unix'] = d.latest_time.apply(lambda t: int(t.timestamp()))
    d['lt_unix3'] = ((d['lt_unix'] - 299.995) + d['Time(msec)']).astype(int)# d['lt_unix2'] = (d['lt_unix']-299.995)
    d['lt'] = pd.to_datetime(d['lt_unix3'], unit='s')
    eq_data = pd.concat([eq_data,d], axis=0, ignore_index=True)
        
eq_data = eq_data[['lt','Time(msec)','POSF(mm)','POSF.1(mm)','TCMD(%)','TCMD.1(%)','TCMD.2(%)','TCMD.3(%)','TCMD.4(%)','SPEED(1/min)']]
print(eq_data.shape)
eq_data.head()
```

```python
"""
eq : データ加工後のeq_data
"""
eq = eq_data

#カラムの要素の変更
for col in eq.columns[1:]:
    eq[col] = eq[col].astype(float)
print(eq.info())

#カラム名を変更
eq = eq.rename(columns={'TCMD(%)': 'X軸_TCMD(%)', 'TCMD.1(%)': 'Z軸_TCMD(%)','TCMD.2(%)':'B軸_TCMD(%)','TCMD.3(%)':'C軸_TCMD(%)','TCMD.4(%)':'スピンドル_TCMD(%)'})

#カラムの削除
eq = eq.drop(['Time(msec)','POSF(mm)','POSF.1(mm)'],axis=1)
```

### 秋山ver

```python
"""
eq_selected : ステップ60で絞ったeq_data (加工開始時間と終了時間を抽出する際に使用する）

"""
# 一定間隔で取り出して1製品分の加工時間を抽出
step = 60
eq_selected = eq.loc[::step].reset_index(drop=True)
eq_selected = eq_selected.sort_values('lt').reset_index(drop=True)#時系列ごとになってなかったので並び替え


# 時系列順にしてから平均を取る。fh_s→前半の約６分のSPEED平均　sh_s→後半の約１分のSPEED平均
for index, row in eq_selected.iterrows(): 
    start = index
    end = start + 1300
    eq_selected.loc[index, 'fh_s'] = eq_selected.loc[start:end, 'SPEED(1/min)'].mean()
    
for index, row in eq_selected.iterrows(): 
    start = index + 1500
    end = start + 400
    eq_selected.loc[index, 'sh_s'] = eq_selected.loc[start:end, 'SPEED(1/min)'].mean()
```

```python
eq_selected.fillna(0, inplace=True)
eq_selected
```

```python
eqtime_list_s =[]
eqtime_list_g =[]
for time,speed,speed2,fh_ave,sh_ave,Ctcmd in zip(eq_selected['lt'], eq_selected['SPEED(1/min)'],eq_selected['SPEED(1/min)'].iloc[3:],eq_selected['fh_s'],eq_selected['sh_s'],eq_selected['C軸_TCMD(%)'].iloc[1000:]):
    if (speed > 10.0) & (speed <= 2500.0) & (speed2 >= 3000.0) & (fh_ave > 3300.0) & (Ctcmd < 0):
        eqtime_list_s.append(time)#加工開始
        if sh_ave >= 3000.0:
            eqtime_list_g.append(time + datetime.timedelta(minutes=9,seconds=55))#910加工終了
        elif sh_ave < 3000.0:
            eqtime_list_g.append(time + datetime.timedelta(minutes=6,seconds=58))#1214加工終了

print('1製品毎の加工時間を抽出完了')
# print(time_list_s)

#抽出した加工時間をデータフレーム化
eqtime_list_s2 = []
eqtime_list_g2 = []
eqtime_list = []
for s in eqtime_list_s:
    for g in eqtime_list_g:
        t = g - s
        if (s not in eqtime_list_s2) & (g not in eqtime_list_g2):
            eqtime_list_s2.append(s)
            eqtime_list_g2.append(g)
            eqtime_list.append(t)
eq_s = pd.DataFrame(eqtime_list_s2, columns = ['加工開始時間'])
eq_g = pd.DataFrame(eqtime_list_g2, columns = ['加工終了時間'])
eq_t = pd.DataFrame(eqtime_list, columns = ['経過時間'])
eq_processing_time = pd.concat([eq_s, eq_g, eq_t], axis=1)

pd.set_option('display.max_rows', None)#作成したデータフレーム結果をすべて表示する#以降すべてに反映されるので注意！
pd.set_option('display.max_columns', None)#作成したデータフレーム結果をすべて表示する#以降すべてに反映されるので注意！
eq_processing_time
```

# time_data加工

```python
"""
time_data: As-prepared time data
"""
#time_data加工
files2 = glob.glob(Timedata_Path)
df_list = [ pd.read_csv(file,encoding="cp932") for file in files2]
time_data = pd.concat(df_list, axis = 0, ignore_index=True)
print(time_data.shape)
time_data.head()
```

```python
### LOT情報取り出し
LOT_LIST = time_data['LotNo'].unique()
LOT_LIST = LOT_LIST.tolist()
AZUL910 = ['0000000-000','2217179-000','2218571-000','2217682-000']
AZUL1214 = ['0000000-000','2216584-000','2217395-000','2215700-000','2216857-000','2217343-000','2217034-000']

#### 1214
LOT_LIST_1214 = [i for i in LOT_LIST if i not in AZUL910]
LOT_LIST_1214

#### 910
# LOT_LIST_910 = [i for i in LOT_LIST if i not in AZUL1214]
# LOT_LIST_910
```

```python
"""
td : データクレンジング後のtimedata
td_0 : tdをフラグ0で絞り込み
td_12 : tdをフラグ1 & 2で絞り込み
"""
td = time_data
# 時間をdatatime型に変換
td["開始時間"] = pd.to_datetime(td["開始時間"])
td["終了時間"] = pd.to_datetime(td["終了時間"])
#クレンジング
td = td.loc[(td["FLAG"] == 0)|(td["FLAG"] == 1)|(td["FLAG"] == 2)]
td.head()
```

# eq_dataとtimedataの結合

```python
for LOT in LOT_LIST_1214:    
    #FLAGで絞り込み
    td_0 = td.loc[(td["FLAG"] == 0) & (td["LotNo"] == LOT)]
    #期間を設定
    START = td_0.iloc[0]["開始時間"]
    FINISH = td_0.iloc[-1]["終了時間"]
    print(LOT)
    print(START)
    print(FINISH)
    
    # td_0 = td_0.loc[(td_0['開始時間'] >= START) & (td_0['開始時間'] <= FINISH)]
    td_12 = td.loc[(td["FLAG"] == 1)|(td["FLAG"] == 2)]
    td_12 = td_12.loc[(td_12['開始時間'] >= START) & (td_12['開始時間'] <= FINISH)]
    # print(td_0.info())
   

    """
    df_time_start : timedataの加工開始時間
    df_eq_start : eqdataの加工開始時間
    df_comb_start : timedataの加工開始時間とeqdataの加工開始時間の結合したデータフレーム
    """
    time_start = []
    eq_start = []

    for a in td_0['開始時間']:
        for b in eq_processing_time['加工開始時間']:
            if (b < ( a + datetime.timedelta(minutes= 10 ))) and (b > ( a + datetime.timedelta(minutes= -10 ))):
                if (a not in time_start) & (b not in eq_start):
                    time_start.append(a)
                    eq_start.append(b)
            continue


    df_time_start = pd.DataFrame(time_start, columns = ['開始時間'])
    df_eq_start = pd.DataFrame(eq_start, columns = ['加工開始時間'])
    df_comb_start = pd.concat([df_time_start, df_eq_start], axis=1)
    print(df_comb_start.head())
    df_comb_start
    
    
    """
    df_comb : timedataとeqdataを結合したデータフレーム
    """
    df_comb = pd.merge(df_comb_start, td_0, how = "outer", on = ["開始時間"])
    # print(df_comb)
    df_comb["加工開始時間"] = pd.to_datetime(df_comb["加工開始時間"])
    # print(df_comb)
    df_comb = pd.merge(df_comb, eq_processing_time, how = "left", on = ["加工開始時間"])
    # print(df_comb)
    df_comb = pd.concat([df_comb,td_12], axis=0, ignore_index=True)

    df_comb = df_comb.rename(columns={'開始時間': 'time_加工開始時間', '加工開始時間': 'eq_加工開始時間','終了時間': 'time_加工終了時間','加工終了時間':'eq_加工終了時間'})
    df_comb = df_comb[['LotNo','FLAG','time_加工開始時間', 'time_加工終了時間','eq_加工開始時間','eq_加工終了時間','経過時間']]
    df_comb = df_comb.sort_values(['time_加工開始時間']).reset_index(drop=True)

    # print(df_comb.shape)
    
    
    #excel出力
    df_comb.to_csv(output_dir + LOT + ".csv",index=False,encoding="cp932")
```
