```python
Path = (r"C:\Users\00220401626\Desktop\910のみtimedata改 (1).xlsx")
df_time = pd.read_excel(Path)
#カラムの結合
df_time['No'] = df_time['No'].astype(str)
df_time["ナンバリング"]= df_time['LotNo']+"-"+ df_time['No']

df_time = df_time.replace("Null",np.nan)
df_time = df_time.dropna()
df_time = df_time.drop(['LotNo','No'],axis=1)
first_column = df_time.pop('ナンバリング')
df_time.insert(0,'ナンバリング',first_column)
d=pd.DataFrame()
for root, dirs, files in os.walk(r"C:\Users\00220401626\Desktop\General\05_グライディングセンタ\05_グライディングセンタ\time_data"):
    for f in files:
        full = os.path.join(root, f)
        di=pd.read_csv(full,encoding="cp932")
        d=pd.concat([d,di], axis=0, ignore_index=True)
dF1=d.loc[d["FLAG"] == 1]
dF2=d.loc[d["FLAG"] == 2]
d=pd.concat([dF1,dF2], axis=0, ignore_index=True)# 時間をdatatime型に変換
d["開始時間"] = pd.to_datetime(d["開始時間"])
d["終了時間"] = pd.to_datetime(d["終了時間"])## 各データのデータ型を確認する。
d.dtypes.to_frame().rename(columns={0:'Column type'})#FLAG=0と結合
data=pd.concat([df_time,d],axis=0)#ソート
data = data.sort_values(by=["開始時間"], ascending=True)#ドレ前
dre_mae = data.loc[(data["FLAG"].shift(-1)==1)&(data["FLAG"]==0),:]#ドレ後
dre_ato = data.loc[(data["FLAG"].shift(2)==2)&(data["FLAG"]==0)&(data["FLAG"].shift(1) != 2),:]##ドレくっつけてソート
dre=pd.concat([dre_mae,dre_ato])
dre["開始時間"]=pd.to_datetime(dre["開始時間"])
dre["終了時間"]=pd.to_datetime(dre["終了時間"])
dre.sort_values(by ="開始時間", ascending = True, inplace = True)
dre.reset_index(inplace=True, drop=True)
dre = dre.drop(['LotNo'],axis=1)
dre
```

```python
import fitz
import pandas as pd
import datetime
import glob, os
from pathlib import Path

folder = Path(r"C:\Users\00220401626\Desktop\General\05_グライディングセンタ\05_グライディングセンタ\三次元測定結果")
files1 = list(folder.glob("*.")


d=pd.DataFrame(columns=["ロットネーム","真直度1(1)","円錐角度_円錐2"])
    
for file in files1:
        files = list(file.glob("*.pdf"))
        doc = fitz.open(file)                                  #ファイル名を取得
        name = file.stem                                       # １ページずつテキストを抽出して出力       
        for page in range(len(doc)):
            text = doc[page].get_text()#リスト化    
            Text = text.split('\n')                             #PDFの中身を抽出
        for i in Text:
            if i =="円錐角度_円錐2":
                Ti2=Text.index("円錐角度_円錐2")
                data2=Text[Ti2+4]
                Ti1=Text.index("真直度1(1)")
                data1=Text[Ti1+5]
                df = pd.DataFrame([[name,data1,data2]],columns=["ロットネーム","真直度1(1)","円錐角度_円錐2"])
                d=pd.concat([d,df],axis=0,ignore_index=True)
d      
```
