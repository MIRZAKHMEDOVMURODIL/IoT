```python
import datetime
import os
import shutil
import glob
import pandas as pd
import numpy as np
import mojimoji
```

```python
#取得範囲
start_time = "2023-02-07"
end_time = "2023-02-08"
start_time = np.datetime64(str(start_time))
end_time = np.datetime64(str(end_time))
#ファイル抽出が済んでいる場合True
skip_sv = False
#ファイル結合が済んでいる場合True
skip_concat = False
```

```python
#サーボビューア抽出用
def read_sv(input_dir,output_dir,st_time,ed_time):
    files = glob.glob(input_dir)
    print(str(len(files))+"files")
    print("処理開始")
    for file,i in zip(files,range(len(files))):
        #出力ファイル名:5桁の連番で出力
        out_file = output_dir+str(str(i).zfill(5))+".csv"
        t = os.path.getmtime(file)
        d = datetime.datetime.fromtimestamp(t)
        if (np.datetime64(d)>=st_time) and (np.datetime64(d)<=ed_time):
            df = pd.read_csv(file,header=[2,3],index_col=0)
            #"file_end_time":ファイルの終了時間
            df.insert(loc = 0, column= "file_end_time", value= str(d))
            #カラム名設定(変更可)
            df.columns = ["file_end_time","Time_msec","X_POS_mm","Z_POS_mm","X_TCMD_%","Z_TCMD_%","B_TCMD_%","C_TCMD_%","S_TCMD_%","S_SPEED_rpm"]
            #"file_start_time":ファイルの開始時間
            df.insert(loc = 0, column= "file_start_time", value= str(d-(pd.to_timedelta(df["Time_msec"].astype(int).max(),"S"))/1000))
            #回転数指定
            df=df[df["S_SPEED_rpm"]>=3490.0]
            df.to_csv(out_file, header=True, index=False, encoding="cp932")
        else:
            continue
    print("出力完了")
```

```python
#元データディレクトリ
in_dir = "**/**/*.csv"
#出力先(一時保存フォルダ)
ot_dir = "tmp/"
#出力先をリセット
shutil.rmtree(ot_dir)
os.mkdir(ot_dir)
if skip_sv == True:
    print("処理をスキップ")
else:
    read_sv(in_dir,ot_dir,start_time,end_time)
    
```

```python
#サーボビューア結合用
def concat_sv(input_dir,output_file):
    files = glob.glob(input_dir)
    first = True
    print(str(len(files))+"files")
    print("処理開始")
    for file in files:
        df = pd.read_csv(file,encoding="cp932")
        if first == True:
            df_all = df.copy()
            first = False
        else:
            df_all = pd.concat([df_all,df],axis=0)
    df_all.to_csv(output_file, header=True, index=False, encoding="cp932")
```

```python
if skip_sv == True:
    print("処理をスキップ")
else:
    concat_sv("tmp/*.csv","all_sv.csv")
```

```python
#まとめたデータを読み込み(上の結合まで済んでいる場合ここから開始)
df_servo = pd.read_csv("all_sv.csv",encoding="cp932")
```

```python
#DATETIMEを作成
df_servo["DATETIME"] = pd.to_datetime(df_servo["file_start_time"])+pd.to_timedelta(df_servo["Time_msec"]/1000,"S")
```

```python
#ユニークid付与
count = 0
uid = 0
time_list = df_servo["DATETIME"]
uid_list = []
th_time = np.timedelta64(60, 's')#判定用閾値 60秒離れていたらプログラムが終了していると判定
for i in time_list:
    if count == 0:
        uid_list.append(uid)
        count += 1
        j = i
    else:
        if i > j + th_time:#1個前のデータからth_time以上離れていたらuidをカウントアップ
            uid += 1
            uid_list.append(uid)
            j = i
        else:
            uid_list.append(uid)
            j = i
```

```python
df_servo["uid"] = uid_list
```

```python
df_uid = df_servo.loc[:,["uid","DATETIME"]].drop_duplicates(subset = "uid")
```

```python
df_std = df_servo.groupby("uid").std()
df_std.reset_index(inplace=True)
```

```python
df_std["FLAG"] = 0
df_std["FLAG"][df_std["B_TCMD_%"]<=0.2] = 1 #B軸の標準偏差が0.2以下はドレッシング
```

```python
df_flag = df_uid.merge(df_std.loc[:,["uid","FLAG"]],how="inner",on="uid")
```

```python
df_flag["lotno"] = "0000000-000"
for k in range(len(df_flag)):
    for i in range(len(df_dress)):
        fdt = np.datetime64(str(df_flag["DATETIME"][k]))
        sdt = np.datetime64(str(df_dress["s1_開始dt"][i]))
        edt = np.datetime64(str(df_dress["s1_終了dt"][i]))
        if (fdt >= sdt ) and (fdt <= edt ) :
            df_flag["lotno"][k] = df_dress["s1_作業指図書no"][i]
        elif (sdt > fdt ):
            break
```
