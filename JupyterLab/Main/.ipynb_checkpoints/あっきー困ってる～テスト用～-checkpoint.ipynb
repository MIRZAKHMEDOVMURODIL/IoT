```python
# データ加工
import numpy as np
import pandas as pd

# 可視化
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib import rcParams 
plt.rcParams["font.family"] = "MS Gothic"
from matplotlib import colors
import matplotlib.dates as mdates
import seaborn as sns

# その他
import datetime
import openpyxl
import warnings
import os
import glob
import fitz
```

```python
"""
変数の定義
"""

# input path
eqdata_Path = r"C:\Users\user\Desktop\05_グライディングセンタ\新しいフォルダー"#フォルダ指定
Timedata_Path = r"C:\Users\user\Desktop\05_グライディングセンタ\time_data/*.csv"#ファイル(*.csv)指定

#output Path
output_dir = r"C:\Users\user\Desktop\05_グライディングセンタ\\"
LOT='2217034-000'
```

```python
"""
eq_data : As-prerared eq_data
"""
eqPath = []

for root, dirs, files in os.walk(eqdata_Path):
    for f in files:
        # 指定したフォルダ内のファイル情報をすべて取得
        full = os.path.join(root, f)
        eqPath.append(full)
        # print(eqPath)
print('全ファイルパス取得完了')
print('現在、全eq_detaを縦結合しています。')

eq_data = pd.DataFrame()
for p in eqPath:    
    t = os.path.getmtime(p)
    D = datetime.datetime.fromtimestamp(t).strftime('%Y/%m/%d %H:%M:%S.%f')[:-3]#最終更新日時をタイムスタンプ型に変換
    
    # ファイル読み込み、整形
    d = pd.read_csv(p, skiprows=2,encoding="cp932")
    d = d.rename(columns={'Time': 'Time(msec)', 'POSF': 'POSF(mm)','POSF.1': 'POSF.1(mm)','TCMD':'TCMD(%)','TCMD.1':'TCMD.1(%)','TCMD.2':'TCMD.2(%)','TCMD.3':'TCMD.3(%)','TCMD.4':'TCMD.4(%)','SPEED':'SPEED(1/min)'})
    d = d.dropna()
    
    d["latest_time"] = D
    d["latest_time"] = pd.to_datetime(d["latest_time"])#最終更新日時をdatetime型に変換
    d['Time(msec)'] = (d['Time(msec)'].astype(float).astype(int))/1000 # d['Time(msec)'] = d['Time(msec)']/1000
    d['lt_unix'] = d.latest_time.apply(lambda t: int(t.timestamp()))
    d['lt_unix3'] = ((d['lt_unix'] - 299.995) + d['Time(msec)']).astype(int)# d['lt_unix2'] = (d['lt_unix']-299.995)
    d['lt'] = pd.to_datetime(d['lt_unix3'], unit='s')
    eq_data = pd.concat([eq_data,d], axis=0, ignore_index=True)
        
eq_data = eq_data[['lt','Time(msec)','POSF(mm)','POSF.1(mm)','TCMD(%)','TCMD.1(%)','TCMD.2(%)','TCMD.3(%)','TCMD.4(%)','SPEED(1/min)']]
print(eq_data.shape)
eq_data.head()
```

```python
"""
eq : データ加工後のeq_data
"""

eq = eq_data

#カラムの要素の変更
for col in eq.columns[1:]:
    eq[col] = eq[col].astype(float)
print(eq.info())

#カラム名を変更
eq = eq.rename(columns={'TCMD(%)': 'X軸_TCMD(%)', 'TCMD.1(%)': 'Z軸_TCMD(%)','TCMD.2(%)':'B軸_TCMD(%)','TCMD.3(%)':'C軸_TCMD(%)','TCMD.4(%)':'スピンドル_TCMD(%)'})

#カラムの削除
eq = eq.drop(['Time(msec)','POSF(mm)','POSF.1(mm)'],axis=1)
```

```python
"""
eq_selected : ステップ60で絞ったeq_data (加工開始時間と終了時間を抽出する際に使用する）

"""
# 一定間隔で取り出して1製品分の加工時間を抽出
step = 60
eq_selected = eq.loc[::step].reset_index(drop=True)
# eq_selected['SPEED(1/min)'].unique()
# #時系列ごとになってなかったので並び替え
eq_selected = eq_selected.sort_values('lt').reset_index(drop=True)

# 時系列順にしてから平均を取る。fh_s→前半の約６分のSPEED平均　sh_s→後半の約１分のSPEED平均    
for index, row in eq_selected.iterrows():
    start = index
    end = start + 1300
    eq_selected.loc[index, 'fh_s'] = eq_selected.loc[start:end, 'SPEED(1/min)'].mean()
    
    start1 = index + 1400
    end2 = start1 + 200
    eq_selected.loc[index, 'sh_s'] = eq_selected.loc[start1:end2, 'SPEED(1/min)'].mean()

print('処理完了しました')
```
