```python
pd.set_option('display.max_rows', None)#作成したデータフレーム結果をすべて表示する#以降すべてに反映されるので注意！
pd.set_option('display.max_columns', None)#作成したデータフレーム結果をすべて表示する#以降すべてに反映されるので注意！

namb = dre17034["s1_ドレスサイクルs"].tolist()


for a,b,c in zip (dre17034['s1_開始dt'],dre17034['s1_終了dt'],namb):
    i=0
    for d in td17034['開始時間']:
        if  (a + (datetime.timedelta(minutes= -10))<d) & (d < (b + datetime.timedelta(minutes= 10 ))):
            td17034.loc[td17034['開始時間'] ==d,"No"] = c + i
            i+=1

td17034
```
