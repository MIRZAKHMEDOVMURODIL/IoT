```python
# 重回帰分析
from statsmodels.formula.api import ols  # statsmodelsライブラリを使用
# + で変数を追加する。
# モデルの学習（crim,zn,indus…を説明変数に使用）
olsmod = ols(formula="mpg ~ cylinders+displacement+horsepower+weight+acceleration+year", data=df).fit()  
print(olsmod.summary())  # 要約の表示
#cylinders+displacement+horsepower+weight+acceleration+year
```

```python
from sklearn import linear_model  # sklearnの回帰モデル
from sklearn.metrics import mean_squared_error  # MSE計算
from sklearn.metrics import r2_score  # 決定係数

# ホールドアウト検証
from sklearn.model_selection import train_test_split
# 説明変数と目的変数に分ける。
X_set = df.drop(['mpg','name','origin'], axis=1)  # 説明変数：medv以外が説明変数 
y_set = df['mpg']  # 目的変数 
# データを分割する。
X_train, X_test, y_train, y_test = train_test_split(X_set, y_set, test_size=0.2, random_state=42)

# モデルの学習
model = linear_model.LinearRegression()
model.fit(X_train, y_train)

# 訓練データの予測値とMSEの計算
y_pred_train = model.predict(X_train)
train_rmse = np.sqrt(mean_squared_error(y_train, y_pred_train))
# テストデータの予測値とMSEの計算
y_pred_test = model.predict(X_test)
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))

# 訓練データでのRMSE
print('訓練RMSE: {}'.format(train_rmse))
# テストデータでのRMSE
print('テストRMSE: {}'.format(test_rmse))
```

```python
# K分割交差検証
from sklearn.model_selection import KFold

# 説明変数と目的変数に分ける。データは先ほどの標準化したBoston住宅データを使う。
X_set = df.drop(['mpg','name','origin'], axis=1)  # 説明変数：medv以外が説明変数 
y_set = df['mpg']  # 目的変数 

# データの分割
kf = KFold(n_splits=5, shuffle=True)

valid_rmse = []
for train_id, test_id in kf.split(X_set):
    X_train = X_set.iloc[train_id]  # 訓練データのidを渡している
    y_train = y_set.iloc[train_id]
    X_test = X_set.iloc[test_id]
    y_test = y_set.iloc[test_id]

   # モデルの学習
    model = linear_model.LinearRegression()
    model.fit(X_train, y_train)

    # テストデータの予測値とRMSEの計算
    y_pred_test = model.predict(X_test)
    test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))
    
    valid_rmse.append(test_rmse)  # 各KのテストMSEの平均がテストRMSE

print('テストRMSE: {}'.format(np.array(valid_rmse).mean()))
```

```python
from sklearn.preprocessing import StandardScaler  # 標準化でつかう
from sklearn.model_selection import train_test_split

# 説明変数と目的変数に分ける。
X_set = df.drop(['mpg','name','origin'], axis=1)  # 説明変数：medv以外が説明変数 
y_set = df['mpg']  # 目的変数 
# ホールドアウト検証
X_train, X_test, y_train, y_test = train_test_split(X_set, y_set, test_size=0.2, random_state=42)

ss = StandardScaler()  # インスタンス化

# 訓練データの標準化
ss.fit(X_train)  # 平均値と標準偏差を計算
X_train_std = pd.DataFrame(ss.transform(X_train))  # 変換
X_train_std.columns = X_train.columns  # 列名を付ける

# テストデータの標準化
X_test_std = pd.DataFrame(ss.transform(X_test))  # 変換
X_test_std.columns = X_test.columns  # 列名を付ける

display(X_train_std.describe())  # 標準化後の説明変数を表示
```

```python
from sklearn.linear_model import LassoCV  # 交差検証で自動でλを選んだくれる
from sklearn.metrics import mean_squared_error

# Lasso
# alphasはλのこと、cvはk分割交差検証のkのこと(ここでやっているのは、λを決めるための分割公差検証）
reg_lasso = LassoCV(alphas=10 ** np.arange(-4, 5, 0.1), cv=5)

#ホールドアウト検証
reg_lasso.fit(X_train_std, y_train)  # 標準化した説明変数を使う

# RMSEを計算
y_train_pred = reg_lasso.predict(X_train_std)  # 予測
rmse_lasso = np.sqrt(mean_squared_error(y_train, y_train_pred))

print('Lasso alpha:{}'.format(round(reg_lasso.alpha_, 4)))   # λを表示
print('Lasso RMSE:{}'.format(round(rmse_lasso,2)))  # RMSEを表示
# 回帰係数
print('係数', reg_lasso.coef_)
# 切片
print('切片', reg_lasso.intercept_)
```

```python
#決定木分析
from sklearn.tree import DecisionTreeRegressor  
# 目的変数がカテゴリの場合（分類問題の場合）はDecisionTreeClassifierを使う。
from sklearn.tree import plot_tree  # 結果の表示用

# 木の深さが3の木をつくる
reg_DT = DecisionTreeRegressor(max_depth=3)  # インスタンス化
reg_DT.fit(X_train, y_train)  # 学習

# 決定木の可視化
plt.figure(figsize=(15, 12))
plot_tree(reg_DT, feature_names=X_train.columns, filled=True)
plt.show()

# 予測値の計算
y_DT_pred = reg_DT.predict(X_test)
# 決定係数の計算
test_r2score = r2_score(y_true=y_test, y_pred=y_DT_pred)
# RMSEの計算(MSEのルートをとっている)
test_rmse = np.sqrt(mean_squared_error(y_true=y_test, y_pred=y_DT_pred))
# 訓練データでのRMSE
print("決定係数: {}".format(test_r2score))
print("RMSE: {}".format(test_rmse))
```

```python
# ランダムフォレスト（回帰）モデルを作る。
from sklearn.ensemble import RandomForestRegressor
reg_RF = RandomForestRegressor(random_state=42)
reg_RF.fit(X_train, y_train)

# 予測値の計算
y_RF_pred = reg_RF.predict(X_test)
# 決定係数の計算
RF_r2score = r2_score(y_true=y_test, y_pred=y_RF_pred)
# RMSEの計算(MSEのルートをとっている)
RF_rmse = np.sqrt(mean_squared_error(y_true=y_test, y_pred=y_RF_pred))
# 訓練データでのRMSE
print("決定係数: {}".format(RF_r2score))
print("RMSE: {}".format(RF_rmse))

# 重要度
features = X_train.columns
importances = reg_RF.feature_importances_  # 重要度を取得
# 重要度のデータフレームの体裁を整える
indices = np.argsort(importances)
RF_importances = pd.DataFrame([features[indices], importances[indices]]).T
RF_importances.columns = ['columns', 'values']
RF_importances = RF_importances.sort_values(by='values', ascending=False)
# 可視化
fig, ax = plt.subplots()
ax.barh(RF_importances['columns'], RF_importances['values'], tick_label=RF_importances['columns'])
```

```python
import lightgbm as lgb  # ligthGBMのインポート
# インスタンス化　引数にlightGBMのパラメータを入れる。
reg_lgb = lgb.LGBMRegressor(objective="regression",   # 回帰問題の場合
                            boosting_type='gbdt',   # ブースティングのタイプ
                            learning_rate='0.1',   # 学習率
                            num_leaves=21,   # 1本の木の最大葉枚数
                            n_estimators=150,  # 作る木の数
                            importance_type="gain",  # 重要度の計算方法
                            random_state=42,  # ランダム変数の固定
                            metric='rmse')  # 評価指標
reg_lgb.fit(X_train, y_train)

# 予測値の計算
y_lgb_pred = reg_lgb.predict(X_test)
# 決定係数の計算
lgb_r2score = r2_score(y_true=y_test, y_pred=y_lgb_pred)
# RMSEの計算(MSEのルートをとっている)
lgb_rmse = np.sqrt(mean_squared_error(y_true=y_test, y_pred=y_lgb_pred))
# 訓練データでのRMSE
print("決定係数: {}".format(lgb_r2score))
print("RMSE: {}".format(lgb_rmse))

# 重要度
plt.rcParams['font.family'] = "MS Gothic"
_, ax = plt.subplots(figsize = (6, 5))
lgb.plot_importance(reg_lgb, height = 0.8, ax = ax)
plt.show()
```
