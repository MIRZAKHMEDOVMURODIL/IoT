```python
import datetime
import os
import glob
import pandas as pd

def read_sv(input_dir, output_dir):
    files = glob.glob(input_dir)
    print(str(len(files)) + " files")
    print("Processing start")
    for file, i in zip(files, range(len(files))):
        # Output file name: 5 digit consecutive numbers for output
        out_file = output_dir + str(str(i).zfill(5)) + ".csv"
        t = os.path.getmtime(file)
        d = datetime.datetime.fromtimestamp(t)
        df = pd.read_csv(file, header=[2,3], index_col=0)
        # "file_end_time": File end time
        df.insert(loc=0, column="file_end_time", value=str(d))
        # Column name setting (can be changed)
        df.columns = ["file_end_time","Time_msec","X_POS_mm","Z_POS_mm","X_TCMD_%","Z_TCMD_%","B_TCMD_%","C_TCMD_%","S_TCMD_%","S_SPEED_rpm"]
        # "file_start_time": File start time
        df.insert(loc=0, column="file_start_time", value=str(d - (pd.to_timedelta(df["Time_msec"].astype(int).max(), "S")) / 1000))
        # Specify rotation number
        df = df[df["S_SPEED_rpm"] >= 3490.0]
        df.to_csv(out_file, header=True, index=False, encoding="cp932")
    print("Processing completed")

# Original data directory
input_dir = r"C:\Users\00220401626\Desktop\General\05_グライディングセンタ\05_グライディングセンタ\eq_data\eq_data\20221108_091036\1\*.csv"
# Output destination
output_dir = r"C:\Users\00220401626\Desktop\General\05_グライディングセンタ\05_グライディングセンタ\eq_data\eq_data\20221108_091036\tets"
```
