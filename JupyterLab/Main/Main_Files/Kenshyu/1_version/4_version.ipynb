```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import glob
from scipy import signal
plt.rcParams["font.family"] = "MS Gothic"
plt.rcParams["font.size"] = 15
```

# 任意の製品のeq_oneディレクトリ

# eq_one読み込み~フーリエ変換

```python
def Fourier_Transform(eq_one):
    # 読み込み～加工
    files = glob.glob(eq_one)
    for file in files:
        df_one = pd.read_csv(file,encoding="cp932")

        step = 1
        df_selected = df_one.loc[::step]
        print(df_selected.shape)

        data1 = df_selected['X軸_TCMD(%)'].tolist()
        data2 = df_selected['Z軸_TCMD(%)'].tolist()
        data3 = df_selected['B軸_TCMD(%)'].tolist()
        data4 = df_selected['C軸_TCMD(%)'].tolist()
        data5 = df_selected['スピンドル_TCMD(%)'].tolist()
        data6 = df_selected['SPEED(1/min)'].tolist()

        # 元グラフ作成 
        fig1 = plt.figure(figsize = (18,10), facecolor='lightblue')
        plt.suptitle(file[-20:-4],fontsize=20)

        #グラフを描画するsubplot領域を作成。
        ax1 = fig1.add_subplot(3, 2, 1)
        ax2 = fig1.add_subplot(3, 2, 2)
        ax3 = fig1.add_subplot(3, 2, 3)
        ax4 = fig1.add_subplot(3, 2, 4)
        ax5 = fig1.add_subplot(3, 2, 5)
        ax6 = fig1.add_subplot(3, 2, 6)

        #各subplot領域にデータを渡す
        x1 = np.linspace(1,len(data1),len(data1))
        x2 = np.linspace(1,len(data2),len(data2))
        x3 = np.linspace(1,len(data3),len(data3))
        x4 = np.linspace(1,len(data4),len(data4))
        x5 = np.linspace(1,len(data5),len(data5))
        x6 = np.linspace(1,len(data6),len(data6))
        ax1.plot(x1, data1)
        ax2.plot(x2, data2)
        ax3.plot(x3, data3)
        ax4.plot(x4, data4)
        ax5.plot(x5, data5)
        ax6.plot(x6, data6)

        # 凡例表示
        ax1.legend(['X軸_TCMD(%)'],loc = 1) 
        ax2.legend(['Z軸_TCMD(%)'],loc = 1) 
        ax3.legend(['B軸_TCMD(%)'],loc = 1) 
        ax4.legend(['C軸_TCMD(%)'],loc = 1) 
        ax5.legend(['スピンドル_TCMD(%)'],loc = 1) 
        ax6.legend(['SPEED(1/min)'],loc = 1) 


        # フーリエ変換
        fft_data1 = np.fft.fft(data1)
        freq_data1 = np.fft.fftfreq(len(data1), d=1/200)
        fft_data2 = np.fft.fft(data2)
        freq_data2 = np.fft.fftfreq(len(data2), d=1/200)
        fft_data3 = np.fft.fft(data3)
        freq_data3 = np.fft.fftfreq(len(data3), d=1/200)
        fft_data4 = np.fft.fft(data4)
        freq_data4 = np.fft.fftfreq(len(data4), d=1/200)
        fft_data5 = np.fft.fft(data5)
        freq_data5 = np.fft.fftfreq(len(data5), d=1/200)
        fft_data6 = np.fft.fft(data6)
        freq_data6 = np.fft.fftfreq(len(data6), d=1/200)

        # パワースペクトル密度のプロット
        fig2 = plt.figure(figsize = (18,10), facecolor='lightblue')
        plt.suptitle(file[-20:-4],fontsize=20)

        #グラフを描画するsubplot領域を作成。
        ax1 = fig2.add_subplot(3, 2, 1)
        ax2 = fig2.add_subplot(3, 2, 2)
        ax3 = fig2.add_subplot(3, 2, 3)
        ax4 = fig2.add_subplot(3, 2, 4)
        ax5 = fig2.add_subplot(3, 2, 5)
        ax6 = fig2.add_subplot(3, 2, 6)

        #各subplot領域にデータを渡す
        ax1.plot(freq_data1[1:len(data1)//2], np.abs(fft_data1[1:len(data1)//2])**2)
        ax2.plot(freq_data2[1:len(data2)//2], np.abs(fft_data2[1:len(data2)//2])**2)
        ax3.plot(freq_data3[1:len(data3)//2], np.abs(fft_data3[1:len(data3)//2])**2)
        ax4.plot(freq_data4[1:len(data4)//2], np.abs(fft_data4[1:len(data4)//2])**2)
        ax5.plot(freq_data5[1:len(data5)//2], np.abs(fft_data5[1:len(data5)//2])**2)
        ax6.plot(freq_data6[1:len(data6)//2], np.abs(fft_data6[1:len(data6)//2])**2)

        # 凡例表示
        ax1.legend(['X軸_TCMD(%)'],loc = 1) 
        ax2.legend(['Z軸_TCMD(%)'],loc = 1) 
        ax3.legend(['B軸_TCMD(%)'],loc = 1) 
        ax4.legend(['C軸_TCMD(%)'],loc = 1) 
        ax5.legend(['スピンドル_TCMD(%)'],loc = 1) 
        ax6.legend(['SPEED(1/min)'],loc = 1) 

        #各subplotにxラベルを追加
        xl = 'Frequency (Hz)'
        ax1.set_xlabel(xl)
        ax2.set_xlabel(xl)
        ax3.set_xlabel(xl)
        ax4.set_xlabel(xl)
        ax5.set_xlabel(xl)
        ax6.set_xlabel(xl)

        #各subplotにyラベルを追加
        yl ='Power Spectral Density'
        ax1.set_ylabel(yl)
        ax2.set_ylabel(yl)
        ax3.set_ylabel(yl)
        ax4.set_ylabel(yl)
        ax5.set_ylabel(yl)
        ax6.set_ylabel(yl)

        # plt.xlim(16.74,16.78)
        # plt.ylim(0,30000000)
        plt.show()
    return 0
```

# フィルター処理(基本は０～２㎐を取り出す。Z軸のみ16.8㎐付近も取り出す。)
# 逆フーリエ変換

```python
def Inverse_Fourier_transform(eq_one):
    # 読み込み～加工
    files = glob.glob(eq_one)
    for file in files:
        df_one = pd.read_csv(file,encoding="cp932")

        step = 1
        df_selected = df_one.loc[::step]
        print(df_selected.shape)

        data1 = df_selected['X軸_TCMD(%)'].tolist()
        data2 = df_selected['Z軸_TCMD(%)'].tolist()
        data3 = df_selected['B軸_TCMD(%)'].tolist()
        data4 = df_selected['C軸_TCMD(%)'].tolist()

        # フーリエ変換
        fft_data1 = np.fft.fft(data1)
        freq_data1 = np.fft.fftfreq(len(data1), d=1/200)
        fft_data2 = np.fft.fft(data2)
        freq_data2 = np.fft.fftfreq(len(data2), d=1/200)
        fft_data3 = np.fft.fft(data3)
        freq_data3 = np.fft.fftfreq(len(data3), d=1/200)
        fft_data4 = np.fft.fft(data4)
        freq_data4 = np.fft.fftfreq(len(data4), d=1/200)


        # X軸フィルター処理
        filtered_fft_data1 = np.zeros(len(fft_data1), dtype=np.complex64)
        for i in range(len(freq_data1)):
            if (freq_data1[i] > 0 and freq_data1[i] < 2):
                filtered_fft_data1[i] = fft_data1[i]
        # X軸フーリエ逆変換
        filtered_data1 = np.fft.ifft(filtered_fft_data1)



        # Z軸フィルター処理
        filtered_fft_data2 = np.zeros(len(fft_data2), dtype=np.complex64)
        for i in range(len(freq_data2)):
            if (freq_data2[i] > 0 and freq_data2[i] < 2)|(16.7< freq_data2[i] < 16.9):
                filtered_fft_data2[i] = fft_data2[i]
        # Z軸フーリエ逆変換
        filtered_data2 = np.fft.ifft(filtered_fft_data2)



        # B軸フィルター処理
        filtered_fft_data3 = np.zeros(len(fft_data3), dtype=np.complex64)
        for i in range(len(freq_data3)):
            if (freq_data3[i] > 0 and freq_data3[i] < 2):
                filtered_fft_data3[i] = fft_data3[i]
        # B軸フーリエ逆変換
        filtered_data3 = np.fft.ifft(filtered_fft_data3)



        # C軸フィルター処理
        filtered_fft_data4 = np.zeros(len(fft_data4), dtype=np.complex64)
        for i in range(len(freq_data4)):
            if (freq_data4[i] > 0 and freq_data4[i] < 2):
                filtered_fft_data4[i] = fft_data4[i]
        # C軸フーリエ逆変換
        filtered_data4 = np.fft.ifft(filtered_fft_data4)

        # パワースペクトル密度のプロット
        fig = plt.figure(figsize = (18,10), facecolor='lightblue')
        plt.suptitle(file[-20:-4],fontsize=20)

        #グラフを描画するsubplot領域を作成。
        ax1 = fig.add_subplot(2, 2, 1)
        ax2 = fig.add_subplot(2, 2, 2)
        ax3 = fig.add_subplot(2, 2, 3)
        ax4 = fig.add_subplot(2, 2, 4)

        #各subplot領域にデータを渡す
        x1 = np.linspace(1,len(data1),len(data1))
        x2 = np.linspace(1,len(data2),len(data2))
        x3 = np.linspace(1,len(data3),len(data3))
        x4 = np.linspace(1,len(data4),len(data4))
        ax1.plot(x1, -(filtered_data1.real),color='blue',alpha=0.9,linewidth=2)
        ax2.plot(x2, filtered_data2.real,color='blue',alpha=0.9,linewidth=2)
        ax3.plot(x3, filtered_data3.real,color='blue',alpha=0.9,linewidth=2)
        ax4.plot(x4, filtered_data4.real,color='blue',alpha=0.9,linewidth=2)

        # 凡例表示
        ax1.legend(['X軸_TCMD(%)'],loc = 1) 
        ax2.legend(['Z軸_TCMD(%)'],loc = 1) 
        ax3.legend(['B軸_TCMD(%)'],loc = 1) 
        ax4.legend(['C軸_TCMD(%)'],loc = 1) 

        #各subplotにxラベルを追加
        xl = "Time[/200(s)]"
        ax1.set_xlabel(xl)
        ax2.set_xlabel(xl)
        ax3.set_xlabel(xl)
        ax4.set_xlabel(xl)

        #各subplotにyラベルを追加
        yl ="ピーク強度"
        ax1.set_ylabel(yl)
        ax2.set_ylabel(yl)
        ax3.set_ylabel(yl)
        ax4.set_ylabel(yl)
        
        ax1.set_ylim(-2,2)
        ax3.set_ylim(-1,1)
        plt.show()
    return 0
```
