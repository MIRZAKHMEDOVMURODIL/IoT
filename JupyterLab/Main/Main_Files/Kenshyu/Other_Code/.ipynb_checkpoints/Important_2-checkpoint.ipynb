# グラフ作成 ### 軸範囲は各自調整してください。

## X軸

```python
# X軸全体
a1_deta = da_selected["X軸_TCMD(%)"].tolist()
b1_deta = db_selected["X軸_TCMD(%)"].tolist()

A = np.array(a1_deta)
B = np.array(b1_deta)

plt.plot(A,alpha=0.9)
plt.plot(B,color = "gold",alpha=0.8)
plt.legend(["OK", "NG2"], fontsize=10, loc=2)
plt.ylim(-3,1.5)
plt.show()
```
