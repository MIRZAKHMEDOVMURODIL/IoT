# WinPythonで運用する

WinPythonは、実行環境にZIPを解凍したら使えるツールです。  
主要なライブラリは入っており、解凍するだけで使えるので、現場の複数台のPCにPython環境を構築するのにはお手軽です。  

【参考】  
[【WinPython】使い方・設定まとめ](https://algorithm.joho.info/programming/python/winpython/)  
[【こりゃ便利】WinPythonでPython環境をUSB化しよう！](https://resanaplaza.com/2022/01/24/%E3%80%90%E3%81%93%E3%82%8A%E3%82%83%E4%BE%BF%E5%88%A9%E3%80%91winpython%E3%81%A7python%E7%92%B0%E5%A2%83%E3%82%92usb%E5%8C%96%E3%81%97%E3%82%88%E3%81%86%EF%BC%81/)　←ダウンロードの手順が丁寧で分かりやすい  

## WinPythonのダウンロード

- ここからダウンロードする（詳細は上のリンク↑を見てください）  
[WinPython - Browse Files at SourceForge.net](https://sourceforge.net/projects/winpython/files/)  
※WinPython_x.xx の x.xx は Python のバージョンではなく、WinPythonのバージョンなので注意  
　例）WinPython_3.10 を選ぶと Python3.9.8 の環境が作成される（Python3.10ではない）  
 ※実行環境で、最低1GBの空きは見ておいてください。空きが少ないから最低限の環境にしたい方は、上記のサイトを参考に、末尾が「dot.exe」になっているものをダウンロードしてください。  
　 ただしその後の設定は自分で頑張ってください（調べたら出てくると思います）  

- ZIPファイルを解凍して使う  
ダウンロードしたZIPファイルを、実行環境へ持っていって解凍する。  

- 実行環境で社外ネットワークが使えない場合や、すごい台数のPCに展開したい（1台ずつpip installしてられない）場合  
あらかじめ**自分のPCで一度解凍＆下記を参照して必要なライブラリをインストールした後、WinPythonのフォルダごとUSBやNASにコピー  
⇒実行環境にコピー**して使用する

※いつも使っているCMDやJupyterlabではないので注意

## WinPythonへライブラリのインストール　　
- **pymysql（DBに接続する系のライブラリ）とpymsteams（Teams送信に使うライブラリ）はインストールが必要**な可能性大  

- CMDからインストールするのが面倒なので、Jupyterlabからインストールする方が楽です。  
 ①上記を参考に、**WinPythonの**Jupyterlabを起動する  
 ②新しいNotebookを作り、セルに`! pip install pymsteams`などのコマンドを実行する    
 
  

- インストールしたライブラリがWinPythonの下にあるか確認する  
 `! pip show ライブラリ名`で確認。自分がダウンロードしたので、Locationのアドレスを見たら分かるはずである  
 
  

- いちいち一つずつインストールするのがめんどくさいという方は、Winpythonに限らず、requirements.txtの使用をご検討ください。  
参考：[Python, pipでrequirements.txtを使ってパッケージ一括インストール](https://note.nkmk.me/python-pip-install-requirements/)

### 外部Webサイトに接続出来ない人（プロキシ・Prisma設定がうまく行かない人）

- 定期実行環境のPCはADドメイン設定がしていない場合があり、外部ネットワークに繋げずインストールできないことがある  
→最初に自分のPCで、上記の手順に従って、必要なライブラリをWinPython以下にインストールしておく。  
　ダウンロード後のWinPythonをフォルダごとUSBにコピーして、実行環境のPCに移す  
→もしくは、実行環境で下記のコードを実行してプロキシ設定してから、インストールのコードを実行する（外部接続できるPCのみ）  
 
```set HTTP_PROXY=http://社員番号:パスワード@10.1.8.72:8080```  
```set HTTPS_PROXY=http://社員番号:パスワード@10.1.8.72:8080```  

※パスワードはプロキシのパスワード。一時的に自分のPWを使用する

## 【おまけ】WinPythonのCMDを使う方法（Jupyter以外で実行したい人向け）
  WinpythonのCMDがあまり言うことを聞いてくれないので、インストールや環境の設定などは、Jupyterlabから「`!`」コマンドを使う方が簡単だと思います。  
  Jupyter使いたくない人向けに、一応記載しておきます。  

### CMDからライブラリインストールする方法
- CMDはいつものPCのCMDではなく、WinPythonのCMDを立ち上げてください。
- 普通にライブラリインストールすると、なぜかデフォルトのPythonフォルダにインストールされてしまいます。  
　→場所指定してインストールしてください。  
　`py -m pip install インストールしたいライブラリ名 -t WinPython 「scripts」フォルダの絶対パス（例　C:\Users\users\Documents\WPy64-3950_winpython\scripts）`  
参考： [【Python】ディレクトリを指定してライブラリをインストールする](https://mgmgblog.com/post-3338/)  

- その他、WinPythonのCMDコマンドはこちら。  
　WinpythonのCMDでは「pip ○○」ではなく、「py -m」を付けないと動かないっぽいですが、JupyterLab上では「pip ○○」じゃないと動きませんでした。謎。  
　[WinPythonにライブラリをpipでインストールする](WinPythonにライブラリをpipでインストールする)  

### ライブラリインストール先のディレクトリがわからない人  
- **WinPythonの**Jupyterlabを起動し、  
　`! pip show pandas`  
　など、既に入っているライブラリのpipを見たら場所が確認できると思います。  

※Winpython解凍フォルダ直下にscriptsフォルダがあるので、おそらくそこ  
※普通のCMDでpip showしても、おそらくWinPython下の環境を拾いません。**WinPythonの**Jupyterlabから実行してください  

- WinPythonのCMDから上記のコマンドを打っても見えると思います（その場合はCMDの初期ディレクトリがWinPythonのscriptsフォルダになっているのを確認）。  

### CMDからWinPythonを実行する
- 下記のコードをCMDで実行  
```python フルパスのファイル名.py```  

※普通に調べると、`python ファイル名のみ.py`というコードがヒットしますが、ファイル名だけでは「そんなファイルはない」というエラーになります。  
　初期フォルダをscriptsフォルダの下だと認識しているようなので、フルパスで指定してやれば動きます。  

- ファイルパスやファイル名にスペースが入ると動かなくなるので、その場合はそもそものファイル・フォルダの名前変更が必要  
スペースを_（アンダーバー）などに置き換えてください。

## 仮想環境下でJupyterlabを起動

- CMDで仮想環境をアクティベートするところまで実行  
[※「CMDに現在アクティブになっている仮想環境名が表示される」までの状態](#anchor)にしておく  

- 仮想環境をアクティベートした状態で、「jupyter lab」を起動  
`jupyter lab`  



- 新しいJupyterの環境（ブラウザ）が立ち上がるので、新しいカーネルのボタンを選んで実行  

**いちいちCMD立ち上げてactivateするのが面倒な人は、batファイルを作る**  
anacondaが使えないので、今のところbatファイルを作成する方法しか分かりません。  
anacondaと同じように、jupyterのconfigファイルに呪文を記載→即起動できる方法を見つけた方は、ご連絡ください。  

- **仮想環境フォルダと同階層のフォルダ**に新しいテキストファイルを作成し、下記3行のコードを入れて保存して閉じる  
`cd /c %~dp0`  
`call 自分が作った環境の名前\Scripts\activate`　★下記のスクショでは、「自分が作った環境の名前」に「testenv」が入っている  
`jupyter lab`

Jupyberlabで仮想環境を使うときは、ショートカットをダブルクリックすると、CMDからアクティベート→jupyterlab立ち上げと同じ動作を自動で実行できる。  

- batファイルを実行すると新しいCMD、ブラウザ、Jupyterlabが立ち上がる。
