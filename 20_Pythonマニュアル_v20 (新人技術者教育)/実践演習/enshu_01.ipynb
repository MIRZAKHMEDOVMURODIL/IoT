````python
df = df_m.copy()

df = df.fillna(0)
df.insert(6, "不良数", df["不良1"] + df["不良2"] + df["不良3"] + df["不良4"] + df["不良5"])
df.insert(6, "良品数", df["検査数"] - df["不良数"])
df
````

````python
# 【別解】
df = df_m.copy()
df = df.fillna(0)
df['不良数'] = df["不良1"] + df["不良2"] + df["不良3"] + df["不良4"] + df["不良5"]
df['良品数'] = df["検査数"] - df["不良数"]

df = df[['LOTNO', '品番', '検査日', '検査者1', '加工元', '検査数', '良品数', '不良数', '不良1', '不良2', '不良3', '不良4', '不良5']]
df
````

````python
# 【おまけ】
# 「不良1～5の行だけをfillnaしたい！他のカラムにNullが入ってて、そっちまでゼロに置き換えたら困る！」という人はこれを使ってください

df1 = df_m.copy()
df1.loc[:, "不良1":"不良5"] = df1.loc[:, "不良1":"不良5"].fillna(0)
df1
````

````python
# 【おまけ】
# 「不良1～5くらいだったらカラム名書くけど、不良1～100だったらどうするの？全部書くのヤダ」という人はこれを使ってください

df1["不良数"] = df1.loc[:, "不良1":"不良5"].sum(axis=1)
df1
````

````python
# カラム名を打ち込むのは大変なので、df.columnsでリスト出力して切り貼りする
df.columns
````
