````python
df = df_m.copy()
df = df[["PAL_1", "PAL_2", "PAL_3"]]

df.insert(0, "LOTNO", file.stem)
df.insert(1, "SNO", df.index + 1)
df.insert(2, "CASENO", (df["SNO"] - 1) // 144 + 1)
df.insert(3, "RNO", (df["SNO"] - 1) % 36 + 1)
df.insert(4, "GNO", (df["SNO"] - 1) // 36 % 4 + 1)
df.insert(5, "UDATE", time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(file.stat().st_mtime)))
df
````

````python
con.execute("truncate fuku03_1")
df_all_m = pd.DataFrame()

for file in files:
    print(file)    # fileは確認用にプリントしている
    df = pd.read_csv(file, encoding="UTF-8")
    df = df[["PAL_1", "PAL_2", "PAL_3"]]

    df.insert(0, "LOTNO", file.stem)
    df.insert(1, "SNO", df.index + 1)
    df.insert(2, "CASENO", (df["SNO"] - 1) // 144 + 1)
    df.insert(3, "RNO", (df["SNO"] - 1) % 36 + 1)
    df.insert(4, "GNO", (df["SNO"] - 1) // 36 % 4 + 1)
    df.insert(5, "UDATE", time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(file.stat().st_mtime)))

# df_allは目視確認用に出している。DBに入れるだけなら、下記のコードは不要
    df_all_m = df_all_m.append(df, ignore_index=True)
#     display(df)
#     display(df_all)

    df.to_sql("fuku03_1", engine, if_exists="append", index=None)
df_all_m
````

````python
df_all = df_all_m.copy()
df_all = df_all.groupby(['LOTNO', 'UDATE'], as_index=False).mean()
df_all = df_all.drop(['SNO', 'CASENO', 'RNO', 'GNO'], axis=1)
df_all = df_all.round({'PAL_1':1, 'PAL_2': 2, 'PAL_3':3})
df_all
````

````python
# con.execute("truncate fuku03_2")
df_all.to_sql("fuku03_2", engine, if_exists="append", index=None)
````
