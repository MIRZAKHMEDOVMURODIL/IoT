````python
import pandas as pd
import sqlalchemy as sa
import pymysql

url = "mysql+pymysql://kenshu:kenshu@10.110.55.59:3306/kenshu?charset=utf8"
engine = sa.create_engine(url)
con = engine.connect()

df1_m = pd.read_sql("select * from kenshu_read.dx_kenshu02_1", engine)
df2_m = pd.read_sql("select * from kenshu_read.dx_kenshu02_2", engine)
df3_m = pd.read_sql("select * from kenshu_read.dx_kenshu02_3", engine)
````

````python
df3_m
````

````python
df1 = df1_m.copy()
df2 = df2_m.copy()
df3 = df3_m.copy()

df1["全LOT数"] = 1
df2["出荷LOT数"] = 1

df4 = pd.merge(df1, df2, on=["拠点", "CASEID", "LOTNO"], how="left")
df4 = df4.groupby(["拠点", "CASEID"], as_index=False).sum()

df5 = pd.merge(df4, df3, on=["拠点", "CASEID"], how="left")
df5 = df5[["拠点", "CASEID", "国内海外", "出荷便", "全LOT数", "出荷LOT数"]]
df5
````

````python
df5.columns
````

````python
# con.execute("truncate fuku02")
df5.to_sql("fuku02", engine, if_exists="append", index=None)
````
