## 特徴
**特徴**
- 汎用性が高い  
  データ連携、データ解析、機械学習、WEBアプリ、RPA、組み込み開発
- コードが非常にシンプル  
  初学者でも比較的容易に扱える  
  共有性が高いため基幹のシステムとしても活用できる  
- （デメリット）C言語と比較して単純な繰り返し処理速度が遅い  
  ※初心者が構築するレベルでは無関係

**支える技術**
- 豊富なライブラリ（拡張機能）  
  ライブラリの開発速度が非常に速い。  
  現時点で、ほとんどの用途に対応した便利なライブラリがある。
- 読みやすさ、書きやすさを意識した文法設計

# 学習のススメ
- 最初に、自身の解決したいテーマ設定を行う事  
  Pythonを学ぶことを目標にしたプログラミングの勉強は、あまり身にならない  
  テーマ設定、及びテーマを実施できる環境が整った後に学ぶ事で、使える技術の形で素早い習得ができる  
  →データエンジニア教育（DE教育）は、自工程のテーマを設定して、そのテーマを実施するのでおすすめです！  
- わからないことをネット検索し解決する能力を鍛える
