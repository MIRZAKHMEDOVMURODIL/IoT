# Pythonファイルの定期実行方法  

## pyファイルの作成  

作成したコードを.py形式で保存する。  
Jupyter lab上から保存形式を変更しても良いが、サクラエディタに貼り付ける方が楽だと思われる。  
 - pyファイルに記述するコードは、定期実行したい内容を記述する。  
  （テーブル定義など、手動で1度だけ行うものは書かない）  
 - ライブラリのimportからエラー処理の終わりまでをセル1つの中にすべて記載し、そのコード（セル）だけで処理が完結するようにする  
 - サクラエディタの場合は、名前を付けて保存するときに拡張子を「.py」と指定すると、pyファイルとして認識されるようになる  

pyファイル用のコードを作ったら、  
・まず、新しいNotebookに、定期実行用のコードを貼り付けて、セル1つだけ実行してうまく行くか確認（エラーが出たら直す）  
・うまく実行できたら、pyファイルにして、pyファイルをダブルクリックして実行（エラーが出たら直す）  
・それもうまくできたら、下記の手順に従い、タスクスケジューラに登録します（動かなければタスクスケジューラの設定がおかしい）  
　※この手順は実行用のPCで行なってください。  
 
【注意】  
「display」や「%pwd」はJupyter用のコマンドのため、pyファイルにするとエラーで止まります。  
display→print, %pwd→import os してos.getcwd() などで置き換えてください。  
 
  

## Windowsタスクスケジューラの設定  

 - 夜中や休日も定期実行したい場合は、常時立ち上げているPCまたはWinSVRなどで実施する。  
 ※タスクスケジューラは、PCの電源が入っていないと動きません  
 - WinSVRを使用する場合、DxにGCPサーバを提供するASUCAというサービスがあります（18,000円/月～）

- 下記を参考に、タスクスケジューラの設定を行う  
https://office54.net/python/app/windows-taskscheduler-exe  
※今回はExeファイルではないので、「プログラムの開始」部分の設定は下記を見て入力  
　（上記と同じサイトです）  
https://office54.net/python/app/windows-taskscheduler-exe#section3-2  

- Python実行用のタスクスケジューラは、新しいフォルダを作った方が、自分のセットしたタスクを探しやすいです。  

**【注意】よく間違えるところ**  
「操作」タブの記載項目は下記のようにする  
・プログラム/スクリプト：　python.exe かpythonw.exeをフルパスで指定  
・引数の追加：　実行するファイル名（○○.py）   
・開始：　上記pyファイルの保存場所（フルパス）  
  

- 「python.exeの場所がどこか分からない」という人は、コマンドプロンプトで下記のコードを打ってください。参考：[Pythonのインストール場所について（Windows）](https://gammasoft.jp/blog/python-install-location/)  
`py --list-paths`  

- タスクスケジューラは「繰り返し間隔」に1と入力すると最短1分間隔で実行できるそうですが、動作確認は各自で行ってください。
