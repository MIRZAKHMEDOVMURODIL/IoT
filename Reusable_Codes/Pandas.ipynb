```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

# Example usage:

for number in countdown(5):
    print(number)
```

```python
# List approach
def create_list_data(n):
    data = []
    for i in range(n):
        data.append(i)
    return data

# Generator approach
def generate_data(n):
    for i in range(n):
        yield i

# Example usage
# Creating a list with 10 million elements
large_list = create_list_data(100000000000000000)

# Generating data using a generator with 10 million elements
large_generator = generate_data(100000000000000000)

# Example usage of the list
print("Sum of elements in large_list:", sum(large_list))

# Example usage of the generator
sum_generator = sum(large_generator)
print("Sum of elements generated by large_generator:", sum_generator)
```
